```python
import pandas as pd
```

```python
df = pd.read_csv("spend.csv")
```

```python
df2=df.loc[df['Agency'] == "63 -  ECONOMIC DEVELOPMENT"]
```

```python
df3 = df2[['Agency','Organization']]
df4 = df3.groupby('Organization').count()
#df4 = df4.sort_index(axis=0)
```

```python
%matplotlib inline
plot = df4.plot.bar()
plot.set_ylabel('Number of entries')
plot.set_xlabel('Organization')
None
```

```python
df5 = df2[['Organization', '2015', '2016','2017']]
df6 = df5.groupby('Organization').sum()
```

```python
%matplotlib inline
plot = df6.plot.bar()
plot.set_yscale('symlog')
plot.set_ylabel('Net spend in fiscal year')
plot.set_xlabel('Organization')
None
```
